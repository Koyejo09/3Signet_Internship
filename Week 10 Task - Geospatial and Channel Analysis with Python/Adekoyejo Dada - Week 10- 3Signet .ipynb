```python
# Create a geopandas Dataframe
geometry = [Point(xy) for xy in zip(df['Latitude'], df['Longitude'])]
gdf = gpd.GeoDataFrame(df, geometry = geometry)
gdf
```

```python
#Display the geopandas DataFrame on a map
gdf.plot(color = 'red')
plt.show()
```

```python
# filter the dataframe to include only longtitude, latitude and transaction amount
df_filtered =df[['Longitude', 'Latitude', 'Sales']]
df_filtered.head(2)
```

```python
#Create a scatter plot to visualize the relationship between longtitude, latitude and sales
plt.figure(figsize = (10,8))
plt.scatter(df_filtered['Longitude'], df_filtered['Latitude'], c =df_filtered['Sales'], 
           cmap = 'viridis')
plt.colorbar(label = 'Sales')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Sales by Latitude and Longtitude')
```

# 2. ANALYSIS OF SALES PERFORMANCE ACROSS DIFFERENT CHANNELS

```python
channel_sales = df.groupby(['Channel'])['Sales'].sum().reset_index()

print(channel_sales)
```

```python
top_channel_sales = channel_sales.sort_values(by='Sales', ascending = False).head(3)
print (top_channel_sales)
```

```python
df = pd.DataFrame(top_channel_sales)
```

```python
plt.figure(figsize=(10,6))
bars = plt.bar(df['Channel'], df['Sales'], color ='skyblue')
# Add data labels on top of each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of the bar (OrderQuantity)
    plt.text(
        bar.get_x() + bar.get_width()/2,  # X position: center of the bar
        yval,  # Y position: height of the bar
        int(yval),  # The actual value to be displayed as text (integer value of OrderQuantity)
        ha='center',  # Horizontal alignment
        va='bottom',  # Vertical alignment (text placed just above the bar)
        fontsize=10  # Set the font size of the labels
    )

plt.bar(df['Channel'], df['Sales'], color='skyblue')
plt.title('Channel by_Sales', fontsize=16)
plt.xlabel('Channel', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.show()
```
